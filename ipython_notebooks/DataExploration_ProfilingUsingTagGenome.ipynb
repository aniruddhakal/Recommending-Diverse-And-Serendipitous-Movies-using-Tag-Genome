```python
user_movies_d = {}

for uid in range(31, 50):
    condition = ratings_matrix_df[uid].notnull()
    user_movies_d[uid] = ratings_matrix_df.index[condition].tolist()
    
# user_movies_df = pd.DataFrame(user_movies_d)
# user_movies_df
user_movies_d
```

```python
# len(user_movies_d[7])
# user_movie_tags_df = pd.DataFrame(index=user_movies_d[1], columns=range(1, 1129))

# collect tag genome scores for all movies that user X has watched

user_x = 49
user_movie_tags_df = pd.DataFrame()

for movie in user_movies_d[user_x]:
    user_movie_tags_df[movie] = genome_scores_df[genome_scores_df['movieId'] == movie]['relevance'].reset_index(drop=True)

user_movie_tags_df = user_movie_tags_df.T

user_movie_tags_df
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4).fit(user_movie_tags_df.values)
result = kmeans.predict(user_movie_tags_df.values)
result
```
