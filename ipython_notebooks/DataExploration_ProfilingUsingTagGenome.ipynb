```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_base_dir = '../datasets/Movielens/'
data_dir = data_base_dir + 'Movielens Latest/ml-latest/'
data_dir2 = data_base_dir + 'Movielens Latest/ml-latest-small/'

genome_scores = data_dir + 'genome-scores.csv'
genome_tags = data_dir + 'genome-tags.csv'
movies = data_dir + 'movies.csv'
ratings = data_dir + 'ratings.csv'
tags = data_dir + 'tags.csv'
```

```python
# movies_df = pd.read_csv(movies)
# movies_df

ratings_df = pd.read_csv(ratings, usecols=range(3), dtype={'userId':np.int64, 'movieId':np.int64, 'rating':np.float64}, low_memory=False)
ratings_df = ratings_df[ratings_df['userId'].isin(range(1, 10000))]
ratings_df
```

```python
ratings_matrix_df = ratings_df.pivot(index='userId', columns='movieId', values='rating')
# ratings_matrix_df = ratings_matrix_df.T
ratings_matrix_df
```

```python
user_movies_d = {}

for uid in range(1, 10000):
    user_movies_d[uid] = ratings_matrix_df.index[ratings_matrix_df[uid].notnull()].tolist()
    
# user_movies_df = pd.DataFrame(user_movies_d)
# user_movies_df
user_movies_d
```

```python
len(user_movies_d[7])
# user_movie_tags_df = pd.DataFrame(index=user_movies_d[1], columns=range(1, 1129))
user_movie_tags_df = pd.DataFrame()


for movie in user_movies_d[4]:
    user_movie_tags_df[movie] = genome_scores_df[genome_scores_df['movieId'] == movie]['relevance'].reset_index(drop=True)

user_movie_tags_df = user_movie_tags_df.T

user_movie_tags_df
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8).fit(user_movie_tags_df.values)
result = kmeans.predict(user_movie_tags_df.values)
result

plt.clf()
# plt.scatter(result, result)
plt.boxplot(result)
plt.show()
```

```python
np.unique(result).size
new_df = pd.DataFrame(result, columns = ['cluster'])

new_df['movie'] = user_movies_d[4]

# cluster ranking and sorting in ascending order
print((new_df.groupby('cluster')['movie'].count()/new_df.size).sort_values().index.tolist())

print(new_df.size)
new_df.groupby('cluster')['movie'].apply(list)
# new_df.set_index(user_movies_d[4], drop=True, inplace=True)
# kmeans.labels_
# genome_tags_df['tag'].values.size
```
