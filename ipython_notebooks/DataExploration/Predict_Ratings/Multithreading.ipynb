```python
import threading as t
import pandas as pd
import matplotlib.pyplot as plt
import multiprocessing as mp

i = 5

def do_this(arg, count):
    print("count ", count)
#     fig = plt.figure()
    plt.title('Count:' + str(count))
    plt.plot(arg)
    plt.show()
    
plist = list()
for a in range(10):
    
#     p = mp.Process(target=do_this, args=([1,2,3], a,))
    p = t.Thread(target=do_this, args=([1,2,3], i,))
    plist.append(p)
    p.start()

for p in plist:
    p.join()
```

```python
import numpy as np
actual = np.array([1, 2.2, 3.1, 4])
predicted = np.array([1, 2, 3, 4])
mae = np.sum(np.absolute(predicted - actual)) / len(predicted)
mse = np.square(mae)
print(mae)
print(mse)
```

```python
from concurrent.futures import ProcessPoolExecutor
import concurrent
import numpy as np

def get_best_cluster(x):
    return x, len(x)

array = np.array([1, 2, 3, 4])
with concurrent.futures.ProcessPoolExecutor(max_workers=2) as executor:
        futures = list()
        for i in range(5):
            futures.append(executor.submit(get_best_cluster, array))

        for f in futures:
            print('result:', type(f.result()[0]))
```
