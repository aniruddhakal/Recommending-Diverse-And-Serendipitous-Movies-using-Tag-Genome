```python
import threading as t
import pandas as pd
import matplotlib.pyplot as plt
import multiprocessing as mp

i = 5

def do_this(arg, count):
    print("count ", count)
#     fig = plt.figure()
    plt.title('Count:' + str(count))
    plt.plot(arg)
    plt.show()
    
plist = list()
for a in range(10):
    
#     p = mp.Process(target=do_this, args=([1,2,3], a,))
    p = t.Thread(target=do_this, args=([1,2,3], i,))
    plist.append(p)
    p.start()

for p in plist:
    p.join()
```
