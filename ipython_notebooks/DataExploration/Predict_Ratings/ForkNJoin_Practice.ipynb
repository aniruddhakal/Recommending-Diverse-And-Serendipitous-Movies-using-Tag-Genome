```python
import multiprocessing as mp
import matplotlib.pyplot as plt

def do_this(arg, count):
    print("")
    fig = plt.figure()
    plt.title('Count:' + str(count))
    plt.plot(arg)
    plt.show()
    
a = 0
while a < 5:
    a += 1
    p = mp.Process(target=do_this, args=([1,2,3], a,))
    p.start()
    p.join()
```

```python
import threading as t
import matplotlib.pyplot as plt

def do_this(arg, count):
    print("")
    fig = plt.figure()
    plt.title('Count:' + str(count))
    plt.plot(arg)
    plt.show()
    
for a in range(5):
#     p = mp.Process(target=do_this, args=([1,2,3], a,))
    p = t.Thread(target=do_this, args=([1,2,3], a,))
    p.start()
#     p.join()
```
