```python
import numpy as np
import pandas as pd
import copy

data_base_dir = '../../datasets/Movielens/'
data_output_dir = data_base_dir + 'output/'
ranking = data_output_dir + 'ranking_composite1.csv'
column_names = ['old_index', 'S_c', 'movieId', 'watched_movie_id', 'S_u', 'diversity','R_cu', 'C_i', 'composite_score']

ranking_df = pd.read_csv(ranking)
ranking_df.columns = column_names
ranking_df.set_index(ranking_df['movieId'].values, drop=True, inplace=True)
ranking_df.drop(columns='old_index', inplace=True)
ranking_df.drop(columns='movieId', inplace=True)
ranking_df.drop(columns='composite_score', inplace=True)
ranking_df
```

```python
ranking_df2 = copy.deepcopy(ranking_df)
ranking_df2['composite_score'] = ranking_df2['R_cu'] * ranking_df2['diversity'] \
                                        * ranking_df2['S_u'] * ranking_df2['S_c'] * ranking_df2['C_i']

ranking_df['rank(R_cu)'] = ranking_df['R_cu'].rank(method='dense').astype(np.int)
ranking_df['rank(diversity)'] = ranking_df['diversity'].rank(method='dense').astype(np.int)
ranking_df['rank(S_u)'] = ranking_df['S_u'].rank(method='dense').astype(np.int)
ranking_df['rank(S_c)'] = ranking_df['S_c'].rank(method='dense').astype(np.int)

ranking_df['composite_score2'] = ranking_df['R_cu'] * ranking_df['rank(R_cu)'] + ranking_df['diversity'] * ranking_df['rank(diversity)'] \
                                        + ranking_df['S_u'] * ranking_df['rank(S_u)'] + ranking_df['S_c'] * ranking_df['rank(S_c)'] + ranking_df['C_i']

# ranking_df.sort_values(['S_c', 'S_u', 'diversity', 'R_cu', 'C_i'], ascending=False)
# ranking_df['composite_score']
ranking_df.sort_values('composite_score2', ascending=False, inplace=True)
ranking_df
```

```python
# Weighted ranking
Rcu_weight = 0.40
Su_weight = 0.20
div_weight = 0.15
Ci_weight = 0.15

ranking_df3 = copy.deepcopy(ranking_df)
ranking_df3.drop(columns='composite_score2', inplace=True)

# reranking with weighting
ranking_df3['composite_score_weighted'] = Rcu_weight * ranking_df['rank(R_cu)'] + div_weight * ranking_df['rank(diversity)'] \
                                + Su_weight * ranking_df['rank(S_u)'] + Ci_weight * ranking_df['C_i']

ranking_df3 = ranking_df3.sort_values('composite_score_weighted', ascending=False)
ranking_df3
```
