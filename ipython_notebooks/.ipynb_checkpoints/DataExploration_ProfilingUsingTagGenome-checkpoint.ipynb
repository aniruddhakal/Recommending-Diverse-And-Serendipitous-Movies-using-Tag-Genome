```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8).fit(user_movie_tags_df.values)

result = kmeans.predict(user_movie_tags_df.values)
result
```

```python
np.unique(result).size
new_df = pd.DataFrame(result, columns = ['cluster'])

len(user_movies_d[user_x])
new_df['movie'] = user_movies_d[user_x]

# cluster ranking and sorting in ascending order
print((new_df.groupby('cluster')['movie'].count()/new_df.size).sort_values().index.tolist())

print(new_df.size)
clusters = new_df.groupby('cluster')['movie'].apply(list)
clusters
# new_df
# new_df.set_index(user_movies_d[user_x], drop=True, inplace=True)
# kmeans.labels_
# genome_tags_df['tag'].values.size
```

```python
cluster_size_df = pd.DataFrame(index=clusters.index)
# cluster_size_df

def process(x):
    x['size'] = len(clusters[x.name])
    
    return x

cluster_size_df['size'] = None
cluster_size_df

cluster_size_df.apply(lambda x: process(x), axis=1)
```

```python
sizes = cluster_size_df['size'].values
below_mean_clusters_df = cluster_size_df[cluster_size_df['size'] <= cluster_size_df['size'].mean()]
below_mean_clusters_df
```

```python
below_mean_clusters_df['rank'] = below_mean_clusters_df.rank(method='dense')
below_mean_clusters_df
```
