```python
# this is how you select values equal to or below quantile threshold
quantile_threshold = 0.25
all_truths_df = genome_scores_df.ge(genome_scores_df.quantile(q=quantile_threshold), axis=1)

def is_similar(col_index, quantile_threshold):
    """
    Unit testing:
    Confirm whether each column, when looked for truth value equal or below quantile threshold,
    produces the same value as done in 1-liner above this function.
    1-liner:
    all_truths_df = genome_scores_df.le(genome_scores_df.quantile(q=quantile_threshold), axis=1)
    """
    n_column_truths = genome_scores_df[col_index].ge(genome_scores_df[col_index].quantile(q=quantile_threshold))
    
    truths = all_truths_df[col_index] == n_column_truths
    
    return truths[truths == True].size == truths.size

for col in genome_scores_df.columns:
    print(is_similar(col, quantile_threshold))
```
